**--Data Cleaning**

```
import pandas as pd
import numpy as np
df=pd.read_csv('99laptop.csv')
print(df.shape)
df.isnull().sum()
```

```
df.head()
```

```
#Dropping columns that seam less useful
df.drop(columns=['name','img'],inplace=True, errors='ignore')
df.columns
```

```
#Fixing missing values
df['released'] = df['released'].fillna(df['released'].mode()[0])
df['Wifi']=df['Wifi'].fillna('No')
df['hdmi']=df['hdmi'].fillna('No')
df['camera']=df['camera'].fillna('No')
df['bluetooth']=df['bluetooth'].fillna('No')
df['spec_score']=df['spec_score'].fillna(df['spec_score'].median())
```

```
df.drop(columns=['usb'], inplace=True)
```

```
#Converting released to released year
df['released_year']=df['released'].str[-4:].astype(int)
df.drop(columns=['released'], inplace=True)
df.head()
```

```
#converting Yes/No to 0's & 1's
cols=['Wifi','bluetooth','camera','hdmi']
for i in cols:
  df[i]=df[i].apply(lambda x:1 if x!='No' else 0)
```

```
df['display_size']=df['display'].str.extract(r'(\d+\.?\d*)').astype(float)
df.drop(columns=['display'],inplace=True)
```

```
df['processor_brand']=df['processor'].str.split().str[0]
df.drop(columns=['processor'],inplace=True)
```

```
df.head()
```

*Splitting storage into RAM and SSD*

```
df['RAM_GB']=df['storage'].str.extract(r'(\d+)\s*GB').astype(int)
df['SSD_GB']=df['storage'].str.extract(r'(\d+)\s*GB SSD').astype(float)
df.drop(columns=['storage'],inplace=True)
```

```
df['windows']=df['windows'].str.replace('-','').str.strip()
```
